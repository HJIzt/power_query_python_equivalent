# Creating `fundamentals_querier()`

I want a function that gets us to the same starting point as where Power Query leaves us - with the cashflow, balance sheet, and income statement of whatever desired company tickers we input.

It needs to do several things:

1. Take a list of stock tickers
2. For each ticker, query cashflow, balance sheet, and income statement for each ticker
3. Join each company's reports together 
4. Concatenate these (with years matched) together for all companies queried
5. Return this dataframe

```python
import datetime
import numpy as np


def fundamentals_querier(tickers, yearly=True):
    """_summary_

    Args:
        tickers (_type_): the string tickers of desired companies
        yearly (bool, optional): whether to return yearly data (if True) or quarterly (if False). Defaults to True.
    """
    # create the dictionary I'm associating the tickers and their fundamentals data in
    fundamentals_dict = {}
    
    for ticker in tickers:
        # query the three dataframes
        cash_flow = si.get_cash_flow(ticker)
        balance_sheet = si.get_balance_sheet(ticker)
        income_statement = si.get_income_statement(ticker)
        
        # concatenate the above three
        ticker_fundamentals = pd.concat([cash_flow, balance_sheet, income_statement])
        
        # change the long date columns to overall year columns
        ticker_fundamentals.columns = pd.to_datetime(ticker_fundamentals.columns.values).year
        
        # assign name of the company to the data
        ticker_fundamentals["ticker"] = ticker
        
        # add the ticker and fundamentals to the dictionary
        fundamentals_dict[ticker] = ticker_fundamentals
        
    
    # create fundamentals_df, which is all the data from the different tickers, concatenated into one
    fundamentals_df = pd.concat(fundamentals_dict.values())
    
    return fundamentals_df
```

# Sample Case: Checking fundamentals of big tech companies

We're going to do the same analysis here as we did in Excel. It's going to also take some cleaning of the dataframe, but we'll arrive in the same place!

For no reason in particular, I decided to use a series of "big tech" companies as our sample. If you change the tickers input to `big_tech`, the entire notebook still runs through correctly, although some variable and plot names may no longer make much sense. 

```python
# first we query our function above to get our data



big_tech = fundamentals_querier(["aapl", "goog", "msft", "meta", "amzn"])
big_tech
```

This table is in wide format - we've got multiple years of the same variable in each row. 

Our first row is the `investments` variable for AAPL from 2019 to 2022. `yahoo_fin` can only query the last 4 years of releases for a firm (no more, no less), and some have releases for 2022 and others don't. If 2022 is out, we see 2019-2022, and if it isn't, then it's 2018-2021. This is why `aapl` has `NaN` values in the 2018 column, while `meta` has `NaN` values in 2022. 

Depending on what you want to do, you may want to get data into long format, as we did in the Excel section. This is one of those cases. I want to make it so every row is an observation of one variable, for one company, in one year. The Pandas `melt` function does this - you specify the `id_vars` argument as whichever fixed identifiers you want, and then the remaining combinations of other variables get split up. 

This is a bad explanation - it'll be clearer once you can see an example below.

```python
big_tech_melt = pd.melt(big_tech.reset_index(), id_vars = ["ticker", "Breakdown"], var_name = "Year")
big_tech_melt
```

So now we've got an equivalent to the dataframe at the end of the PowerQuery operations. Each row is one company's value of a certain variable (i.e. `Breakdown`) in a certain year.

Next we want to comlete the pivot table. We can't set filters in the Pandas `pivot_table` function, but we actually swapped the `years` variable to the columns anyways when we did our analysis. So, this will work fine for what we need.

```python
# second pivot table version equivalent
# excel has year and company as columns, categories as rows, values summed as values

table_2_equiv = pd.pivot_table(big_tech_melt, values = "value", columns = ["ticker", "Year"], index = "Breakdown", aggfunc = np.sum)
table_2_equiv
```

In order to make calculated columns (equivalent to the calculated fields of Excel) we need to have the variables each be columns. This merely requires we transpose the dataframe, with the .T method.

```python
table_2_transpose = table_2_equiv.copy().T
table_2_transpose
```

```python
# table 2 calculated fields
table_2_calculations = table_2_transpose.copy()
table_2_transpose["CurrentRatio"] = table_2_transpose["totalCurrentAssets"]/table_2_transpose["totalCurrentLiabilities"]
table_2_transpose["DebtRatio"] = table_2_transpose["totalLiab"]/table_2_transpose["totalAssets"]

table_2_transpose["ProfitMargin_OperatingProfit"] = table_2_transpose["operatingIncome"]/table_2_transpose["totalRevenue"]
table_2_transpose["ProfitMargin_NetIncome"] = table_2_transpose["netIncomeApplicableToCommonShares"]/table_2_transpose["totalRevenue"]
table_2_transpose["InterestCoverage"] = table_2_transpose["operatingIncome"]/table_2_transpose["interestExpense"]

table_2_transpose = table_2_transpose[["CurrentRatio", "DebtRatio", "ProfitMargin_OperatingProfit", "ProfitMargin_NetIncome", "InterestCoverage"]]
table_2_transpose
```

I cannot plot this - I need to reset the index and fill the NaN values with 0. 

```python
table_2_transpose_resetindex = table_2_transpose.reset_index()
t2_nanzero = table_2_transpose_resetindex.fillna(0)
t2_nanzero
```

Now it's ready for plotting. I'm going to make a series of grouped bar plots. Each barplot shows one variable for each company and year.
